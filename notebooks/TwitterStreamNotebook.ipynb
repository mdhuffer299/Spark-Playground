```pyspark
tweets = sqlContext.read.json("wasb:///TwitterLogs/TwitterLogs_02_11_2018.json")
```

```pyspark
tweets.count()
```

```pyspark
tweets.registerTempTable("TWEETS")
```

```pyspark
%%sql

SELECT *
FROM TWEETS
LIMIT 10
```

```pyspark
from pyspark.sql.functions import split
```

```pyspark
tweets = tweets.withColumn("text", split("text", "\s+"))
```

```pyspark
tweets.select('text').take(1)
```
