```pyspark
vinDF = spark.read.csv('wasb:///VIN_LOOKUP.csv', header =True , inferSchema = True)
groundedDF = spark.read.csv('wasb:///GROUNDED_VEHICLES.csv', header = True, inferSchema = True)
```

```pyspark
groundedDF.count()
```

```pyspark
saleDF.createTempView('saleDF')
vinDF.createTempView('vinDF')
groundedDF.createTempView('groundedDF')
```

```pyspark
df_1 = saleDF.join(vinDF, saleDF.vin_810 == vinDF.vin, how = 'full')
df_1.take(5)
```

```pyspark
fullDF = df_1.join(groundedDF, df_1.vin == groundedDF.vin, how = 'full')
fullDF.take(5)
```

```pyspark
fullDF.count()
```
